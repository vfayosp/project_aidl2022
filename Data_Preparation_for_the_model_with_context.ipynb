# Defining the preparation functions

### get_indexes function:

This function given a numpy dataframe, the name of the rows and columns and the minimum interactions we want the data to obly to have it returns the indexes of the rows and columns that do not fulfill the requierements.


```python
def get_indexes(data, row_name, col_name, min_row = 2, min_col=1):
  print("___________________________________________________________________")
  print("                        ",row_name,"-",col_name, "                        ")
  print("___________________________________________________________________")
  lrow=list(data.sum(axis=1))
  print(" - Total number of ",row_name+"s: ", len(lrow))
  index_rows = [i for i in range(len(lrow)) if lrow[i] <min_row ]
  print(" - "+row_name+"s with less than",min_row,"interactions:",len(index_rows))
  lcol=list(data.sum(axis=0))
  print(" - Total number of ",col_name+"s: ", len(lcol))
  index_cols = [i for i in range(len(lcol)) if lcol[i] <min_col ]
  print(" - "+col_name+"s with less than",min_col,"interactions:",len(index_cols),"\n")
  return(index_rows,index_cols)

```

### delete_indexes function:

This function given the numpy dataframe, the name of the rows and columns and the indexes of the rows and columns that we want to delete, it deletes these indexes and returns the data completely cleaned updating the dictionary of shapes.

```python
def delete_indexes(data,row_name, col_name, index_rows,index_cols, final_shapes):
  data = np.delete(data, index_rows, axis=0)
  data = np.delete(data, index_cols, axis=1)
  final_shapes[row_name+"_"+col_name]=data.shape
  return(data,final_shapes)
```

### prepare function:
Given the three datasets it iterates till the data is accepted by the standards.

```python
def prepare(data_drug_protein,data_protein_disease,data_drug_disease):
  original_shapes = {"drug_protein": data_drug_protein.shape, "protein_disease": data_protein_disease.shape, "drug_disease": data_drug_disease.shape}
  final_shapes = {"drug_protein": data_drug_protein.shape, "protein_disease": data_protein_disease.shape, "drug_disease": data_drug_disease.shape}
  accepted = False
  iteration = 1
  while not accepted:
    print("*******************************************************************")
    print("                        Iteration "+ str(iteration)+ ":                        ")
    print("*******************************************************************")
    (index_drugs1,index_proteins1)=get_indexes(data_drug_protein, "drug", "protein")
    (index_proteins2,index_diseases1)=get_indexes(data_protein_disease, "protein", "disease")
    (index_drugs2,index_diseases2)=get_indexes(data_drug_disease, "drug", "disease")
    index_drugs = list(set(index_drugs1) | set(index_drugs2))
    index_proteins= list(set(index_proteins1) | set(index_proteins2))
    index_diseases= list(set(index_diseases1) | set(index_diseases2))
    if len(index_drugs)>0 or len(index_proteins)>0 or len(index_diseases)>0:
      (data_drug_protein,final_shapes)=delete_indexes(data_drug_protein,  "drug","protein",index_drugs,index_proteins,final_shapes)
      (data_protein_disease,final_shapes)=delete_indexes(data_protein_disease, "protein", "disease",index_proteins,index_diseases,final_shapes)
      (data_drug_disease,final_shapes)=delete_indexes(data_drug_disease, "drug", "disease",index_drugs,index_diseases,final_shapes)
      iteration+=1
    else:
      accepted = True
      print(" ==> THE DATA IS ACCEPTED \n")
  print(" ===> Changes of shape:")
  for i in original_shapes:
    print("        * "+i+" went from "+ str(original_shapes[i])+" to "+str(final_shapes[i]))
  return(data_drug_protein,data_protein_disease,data_drug_disease)
  


```

# Preparing the data:

```python
(data_drug_protein,data_protein_disease,data_drug_disease)=prepare(data_drug_protein,data_protein_disease,data_drug_disease)
```

# Function to compute and save train and test for each dataset:
### compute_and_save_train_test this function randomly takes one of the columns of each row to use it as test and saves the other ones to use them as train. We will save the data with a 2 in order to distinguish them from the data used for the first part of the project.

```python
def compute_and_save_train_test(data,row_name, col_name,saving_path):
  print("*******************************************************************")
  print("                        ",row_name,"-",col_name, "                        ")
  print("*******************************************************************")
  my_data_train = []
  my_data_test = []
  for i in range(data.shape[0]):
    lista= list(np.where(data[i])[0])
    r = random.choice(lista)
    lista.remove(r)
    my_data_test.append([i, r + data.shape[0],1])
    for j in lista:
      my_data_train.append([i, j + data.shape[0],1])
  df_train = pd.DataFrame(my_data_train, columns = [row_name, col_name, 'value'])
  print("Train:")
  print(df_train.head(3))
  df_test = pd.DataFrame(my_data_test, columns = [row_name, col_name, 'value'])
  print("Test:")
  print(df_test.head(3))
  print("...Saving them... ")
  df_train.to_csv(saving_path+'df_train_'+row_name+'_'+col_name+'2.csv', sep='\t')
  df_test.to_csv(saving_path+'df_test_'+row_name+'_'+col_name+'2.csv', sep='\t')  

  
```
