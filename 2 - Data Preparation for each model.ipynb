# Defining the preparation functions

### get_indexes function:

This function given a numpy dataframe, the name of the rows and columns and the minimum interactions we want the data to obly to have it returns the indexes of the rows and columns that do not fulfill the requierements.


```python
def get_indexes(data, row_name, col_name, min_row = 2, min_col=1):
  print("___________________________________________________________________")
  print("                        ",row_name,"-",col_name, "                        ")
  print("___________________________________________________________________")
  lrow=list(data.sum(axis=1))
  print(" - Total number of ",row_name+"s: ", len(lrow))
  index_rows = [i for i in range(len(lrow)) if lrow[i] <min_row ]
  print(" - "+row_name+"s with less than",min_row,"interactions:",len(index_rows))
  lcol=list(data.sum(axis=0))
  print(" - Total number of ",col_name+"s: ", len(lcol))
  index_cols = [i for i in range(len(lcol)) if lcol[i] <min_col ]
  print(" - "+col_name+"s with less than",min_col,"interactions:",len(index_cols),"\n")
  return(index_rows,index_cols)

```

### delete_indexes function:

This function given the numpy dataframe and the indexes of the rows and columns that we want to delete, it deletes these indexes and returns the data completely cleaned.

```python
def delete_indexes(data,index_rows,index_cols):
  data = np.delete(data, index_rows, axis=0)
  data = np.delete(data, index_cols, axis=1)
  return(data)
```

### prepare function:
Given the three datasets it iterates till the data is accepted by the standards.

```python
def prepare_data(data,row_name, col_name):
 (index_rows,index_columns)=get_indexes(data, row_name, col_name)
 print(" ===> Change of shape:")
 print("       * Original shape: "+str(data.shape))
 data = delete_indexes(data,index_rows,index_columns) 
 print("       * Final shape: "+str(data.shape))                                          
 return(data)
```

# Function to compute and save train and test for each dataset:
### compute_and_save_train_test this function randomly takes one of the columns of each row to use it as test and saves the other ones to use them as train. 

```python
def compute_and_save_train_test(data,row_name, col_name,saving_path):
  print("*******************************************************************")
  print("                        ",row_name,"-",col_name, "                        ")
  print("*******************************************************************")
  my_data_train = []
  my_data_test = []
  for i in range(data.shape[0]):
    lista= list(np.where(data[i])[0])
    r = random.choice(lista)
    lista.remove(r)
    my_data_test.append([i, r + data.shape[0],1])
    for j in lista:
      my_data_train.append([i, j + data.shape[0],1])
  df_train = pd.DataFrame(my_data_train, columns = [row_name, col_name, 'value'])
  print("Train:")
  print(df_train.head(3))
  df_test = pd.DataFrame(my_data_test, columns = [row_name, col_name, 'value'])
  print("Test:")
  print(df_test.head(3))
  print("...Saving them... ")
  df_train.to_csv(saving_path+'df_train_'+row_name+'_'+col_name+'.csv', sep='\t')
  df_test.to_csv(saving_path+'df_test_'+row_name+'_'+col_name+'.csv', sep='\t')  

  
```

# Computing train and test and saving them

```python
saving_path='gdrive/MyDrive/postgraduate_program/project/AIDL_recommended_systems/data/prepared_data/'
```

```python
compute_and_save_train_test(data_drug_protein, "drug", "protein",saving_path)
compute_and_save_train_test(data_protein_drug, "protein", "drug",saving_path)
compute_and_save_train_test(data_protein_disease, "protein", "disease",saving_path)
compute_and_save_train_test(data_drug_disease, "drug", "disease",saving_path)
```
